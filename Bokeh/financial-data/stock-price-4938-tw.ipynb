```python
'''
This time series example is based on the stock price data from
https://finance.yahoo.com/quote/4938.TW/history?p=4938.TW
'''
import pandas as pd
from collections import OrderedDict
from bokeh.charts import TimeSeries, output_notebook, show
```

```python
# load data
stock_data = pd.read_csv("data/4938.TW.csv")
```

```python
# show information of stock_data
print("(total records, dimensionality) =", stock_data.shape)
print(stock_data.head())
```

```python
output_notebook()
```

```python
# compose dict with Adj Close and Date
data = OrderedDict(TW4938=stock_data["Adj Close"], Date=stock_data["Date"])

# create a time series with a title, a legend and axes
p = TimeSeries(data, x="Date", legend=True, title="4938 TW Stock Price", ylabel="Adj Close", color="orange")

# show the 
show(p)
```
