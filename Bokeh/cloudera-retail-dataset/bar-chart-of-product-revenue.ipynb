```python
'''
This bar chart example is based on Cloudera's retail database.
https://www.cloudera.com/downloads/quickstart_vms/5-12.html
'''
import pandas as pd
import matplotlib.pyplot as plt
from bokeh.charts import Bar, output_notebook, show
from bokeh.charts.attributes import CatAttr
```

```python
# prepare dataset
product_data = pd.DataFrame([
    [1004, "Field & Stream Sportsman 16 Gun Fire Safe", 6637668.2823181152],
    [365, "Perfect Fitness Perfect Rip Deck", 4233794.3682899475],
    [957, "Diamondback Women's Serene Classic Comfort Bi", 3946837.0045471191],
    [191, "Nike Men's Free 5.0+ Running Shoe", 3507549.2067337036],
    [502, "Nike Men's Dri-FIT Victory Golf Polo", 3011600],
    [1073, "Pelican Sunstream 100 Kayak", 2967851.6815185547],
    [1014, "O'Brien Men's Neoprene Life Vest", 2765543.314743042],
    [403, "Nike Men's CJ Elite 2 TD Football Cleat", 2763977.4868011475],
    [627, "Under Armour Girls' Toddler Spine Surge Runni", 1214896.220287323],
    [565, "adidas Youth Germany Black/Red Away Match Soc", 63490]
], columns = ["product_id", "product_name", "revenue"]);
```

```python
print("(total products, dimensionality) =", product_data.shape)
print(product_data)
```

```python
# set figure
plt.figure();

# create a bar chart
p = product_data.plot.bar(x="product_name", y="revenue");

# show the results
plt.show();
```

```python
output_notebook()
```

```python
# create a bar chat with a title and axis labels
p = Bar(product_data, values="revenue", label=CatAttr(columns=["product_name"], sort=False), title="Cloudera Retail Product Dataset", xlabel="Product Name", ylabel="Revenue", legend=False, color="navy")

# show the results
show(p)
```
