```python
from bokeh.plotting import figure, output_notebook, show
import numpy as np

output_notebook()
```

```python
# prepare data
x = np.sort(np.round(np.random.rand(6) * 40))
y1 = np.round(np.random.rand(6) * 30)
y2 = np.round(np.random.rand(6) * 30)

# create a plot with a title and axis labels
p = figure(title="random line", x_axis_label='x', y_axis_label='y')

# add a line renderer with legend, line thickness and line color
p.line(x, y1, legend="type 1", line_width=2, line_color='red')
p.line(x, y2, legend="type 2", line_width=2, line_color='blue')

# show the results
show(p)
```
