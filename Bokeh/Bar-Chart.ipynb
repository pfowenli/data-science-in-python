```python
from bokeh.charts import Bar, output_notebook, show
from bokeh.models.ranges import Range1d

output_notebook()
```

```python
# prepare data
data = {
    "bank": ["Mega", "Taiwan", "Taiwan Cooperative", "Chang Hwa"],
    "rate": [29.73, 29.71, 29.795, 29.857]
}

# create a new line chat with a title and axis labels
p = Bar(data, values="rate", label="bank", title="Exchange rate Bar Chart", xlabel='bank', ylabel='exhange rate(TWD/USD)', legend=False, width=400, height=400)

# set y range
p.y_range = Range1d(29, 30)

# show the results
show(p)
```
